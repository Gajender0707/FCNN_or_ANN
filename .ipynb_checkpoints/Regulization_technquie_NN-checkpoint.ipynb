```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("kc_house_data.csv")
```

```python
data.head()
```

```python
data=data.drop(["id","date"],axis=1)
```

```python
data.head()
```

```python
y=data.iloc[:,0:1]
```

```python
y.head()
```

```python
x=data.drop("price",axis=1)
```

```python
x.shape
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
k=StandardScaler()
k.fit(x_train)
x_train=k.transform(x_train)
x_test=k.transform(x_test)
```

```python
k1=StandardScaler()
k1.fit(y_train)
y_train=k1.transform(y_train)
y_test=k1.transform(y_test)
```

```python
x_train
```

```python
y_train
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adagrad
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import RMSprop
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import BatchNormalization
```

# Dropout:  

```python
def baseline_model():
    model=Sequential()
    model.add(Dropout(0.2,input_dim=18))  #dropout is the palenty or the regulization on the data
    
    model.add(Dense(128,activation="relu",input_dim=18))
    model.add(Dropout(0.2))
    model.add(Dense(64,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(32,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(16,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(8,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(4,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    opt=Adam(learning_rate=0.001)
    model.compile(loss="mean_squared_error",optimizer=opt)
    return model
```

```python
model=baseline_model()
```

```python
model.summary()
```

```python
history=model.fit(x_train,y_train,verbose=1,batch_size=64,epochs=200,validation_data=(x_test,y_test))
```

```python
from sklearn.metrics import r2_score
```

```python
yp_train=model.predict(x_train)
```

```python
yp_test=model.predict(x_test)
```

```python
r2_score(y_train,yp_train)
```

```python
r2_score(y_test,yp_test)
```

```python
plt.plot(history.history["loss"],color="green",label="training loss")
plt.plot(history.history["val_loss"],color="red",label="test loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend()
```

```python
#hence the probelm of overfit is the fix by the Dropout 
```

# BatchNormalizer:

```python
def baseline_model():
    model=Sequential()
    model.add(Dense(128,activation="relu",input_dim=18))
    model.add(BatchNormalization())    #BatchNormalization is the technquie to remove overfitting by regulization
    model.add(Dense(64,activation="relu"))
    model.add(BatchNormalization())   #BatchNormalization is the fast in the processing 
    model.add(Dense(32,activation="relu"))
    model.add(BatchNormalization())
    model.add(Dense(16,activation="relu"))
    model.add(BatchNormalization())
    model.add(Dense(8,activation="relu"))
    model.add(BatchNormalization())
    model.add(Dense(4,activation="relu"))
    model.add(BatchNormalization())
    model.add(Dense(1))
    opt=Adam(learning_rate=0.001)
    model.compile(loss="mean_squared_error",optimizer=opt)
    return model
```

```python
model=baseline_model()
```

```python
history=model.fit(x_train,y_train,verbose=1,batch_size=64,epochs=200,validation_data=(x_test,y_test))
```

```python
yp_train=model.predict(x_train)
yp_test=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
r2_score(y_train,yp_train)
```

```python
r2_score(y_test,yp_test)
```

```python
l=np.sqrt(mean_squared_error(y_test,yp_test))
l
```

```python
plt.plot(history.history["loss"],label="training loss")
plt.plot(history.history["val_loss"],label="testing loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend()
```
