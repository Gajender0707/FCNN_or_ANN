```python
y=data.generated_power_kw
```

```python
x=data.drop("generated_power_kw",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/2, random_state = 0)
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adagrad
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import RMSprop
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from tensorflow.keras.layers import LeakyReLU
```

```python
# def baseline_model():
#     model=Sequential()
#     model.add(Dense(30,input_dim=20,activation="relu"))
#     model.add(Dense(20,activation="relu"))
#     model.add(Dense(10,activation="relu"))
#     model.add(Dense(5,activation="relu"))
#     model.add(Dense(1))
#     opt=RMSprop(learning_rate=0.000000001)
#     model.compile(loss="mean_squared_error",optimizer=opt)
#     return model
```

```python
def baseline_model():
    model=Sequential()
    model.add(Dense(30,input_dim=20,activation=LeakyReLU(alpha=0.1)))
    model.add(Dense(20,activation=LeakyReLU(alpha=0.1)))
    model.add(Dense(10,activation=LeakyReLU(alpha=0.1)))
    model.add(Dense(1))
    opt=Adam(learning_rate=0.01)
    model.compile(loss="mean_squared_error",optimizer=opt)
    return model
```

```python
model=baseline_model()
```

```python
model.fit(x_train,y_train,verbose=1,epochs=200,batch_size=1040,validation_data=(x_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
yp_train=model.predict(x_train)
```

```python
np.sqrt(mean_squared_error(yp_train,y_train))
```

```python
r2_score(y_train,yp_train)
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(x_train,y_train)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
y_pred=model1.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
