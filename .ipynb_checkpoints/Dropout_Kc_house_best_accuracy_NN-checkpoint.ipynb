```python
def Cl(data):
    import numpy as np
    
    for i in range(len(data.columns)):
        
        y=data.iloc[:,i]
        out=[]
        mean=np.mean(y)
        std=np.std(y)
        for j in y:
            z=(j-mean)/std
            if(z>3):
                out.append(j)
        for m in out:
            data=data[data.iloc[:,i]!=m]
    return data
data1=Cl(data)
```

```python
def baseline_model():
    model=Sequential()
    model.add(Dense(256,input_dim=16,activation="relu"))
    model.add(Dropout(0.2,input_dim=16))   #dropout is the palenty or the regulization on the data 
    model.add(Dense(128,input_dim=16,activation="relu"))
    model.add(Dropout(0.2,input_dim=16))
    model.add(Dense(64,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(32,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(16,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(8,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(4,activation="relu"))
    model.add(Dropout(0.2))
    model.add(Dense(2,activation="relu"))
    model.add(Dense(1))
    opt=Adam(learning_rate=0.0001)
    model.compile(loss="mean_squared_error",optimizer=opt)
    return model
```

```python
model=baseline_model()
```

```python
history=model.fit(x_train,y_train,verbose=1,batch_size=64,epochs=300,validation_data=(x_test,y_test))
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
yp_train=model.predict(x_train)
yp_test=model.predict(x_test)
```

```python
r2_score(y_train,yp_train)
```

```python
r2_score(y_test,yp_test)
```

```python
plt.plot(history.history["loss"],color="green",label="training loss")
plt.plot(history.history["val_loss"],color="red",label="test loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend()
```
